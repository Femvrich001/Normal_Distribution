```python
Countries= df["Country"]

# create a frequency_table 
frequency_table = pd.DataFrame({
    "Country": Countries.value_counts().index,
    "Absolute_frequency": Countries.value_counts().values,
    "Relative_frequency": (Countries.value_counts(normalize=True).values*100).round(0),
    "Cummulative_frequency": (Countries.value_counts(normalize=True).values*100).round(0)
})
# sort the table in descending order in order of the Absolute _frequency
frequency_table = frequency_table.sort_values(by='Absolute_frequency',ascending=False).reset_index(drop=True)

# create another table to display the total 
total_row = pd.DataFrame({
    'Country': ['Total'],
    'Absolute_frequency': [frequency_table['Absolute_frequency'].sum()],
    'Relative_frequency': [100],
    'Cummulative_frequency': [100]
})

# concatinating the frequency_table and the total_table
frequency_table = pd.concat([frequency_table,total_row], ignore_index=True)

print(frequency_table)

```

```python
# Your frequency table
data = {
    'Country': ['USA', 'Canada', 'Russia', 'UK', 'Belgium', 'Denmark', 'Germany', 'Mexico', 'Total'],
    'Absolute_frequency': [2, 2, 1, 1, 1, 1, 1, 1, 10],
    'Relative_frequency': [20, 20, 10, 10, 10, 10, 10, 10, 100],
    'Cummulative_frequency': [20, 40, 50, 60, 70, 80, 90, 100, 100],
}

frequency_table = pd.DataFrame(data)

# Sort the DataFrame by 'Absolute_frequency' in descending order
frequency_table = frequency_table.sort_values(by='Absolute_frequency', ascending=False)

# Create a Pareto diagram
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar chart
ax1.bar(frequency_table['Country'], frequency_table['Absolute_frequency'], color='b', alpha=0.7, align='center')

# Line chart (Cumulative Percentage)
ax2 = ax1.twinx()
ax2.plot(frequency_table['Country'], frequency_table['Cummulative_frequency'], color='r', marker='o')

# Set labels and title
ax1.set_xlabel('Country')
ax1.set_ylabel('Absolute Frequency', color='b')
ax2.set_ylabel('Cumulative Frequency (%)', color='r')
plt.title('Pareto Diagram')

# Show the plot
plt.show()

```

```python
import numpy as np
from scipy.stats import skew

# Extract the 'Price' column
price_data = df['Price'].dropna()  # Drop missing values, if any

# Mean
mean_price = np.mean(price_data)

# Median
median_price = np.median(price_data)

# Mode
mode_price = price_data.mode()[0]  # Using [0] as there might be multiple modes

# Skewness
skewness_price = skew(price_data)

# Variance
variance_price = np.var(price_data, ddof=1)  # Use ddof=1 for sample variance

# Standard Deviation
std_dev_price = np.std(price_data, ddof=1)  # Use ddof=1 for sample standard deviation

# Print the results
print(f"Mean Price: {mean_price:.2f}")
print(f"Median Price: {median_price:.2f}")
print(f"Mode Price: {mode_price:.2f}")
print(f"Skewness of Price: {skewness_price:.2f}")
print(f"Variance of Price: {variance_price:.2f}")
print(f"Standard Deviation of Price: {std_dev_price:.2f}")


```
