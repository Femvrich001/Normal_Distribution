```python
import pandas as pd
import numpy as np

file_path= r"C:\Users\Olajide FemVrich\Documents\FemVrich\DATA\STATS\Section 3\4_The Standard Normal Distribution\Exercises\Normal_distribution_exercise.xlsx"
sheet_name = 0
df = pd.read_excel(file_path, sheet_name=sheet_name, skiprows=9, usecols=[1])
df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Ensure inline display of figures in Jupyter notebook
%matplotlib inline

# Assuming df is the DataFrame containing your data
data_column = df['Original dataset']

# Calculate mean and standard deviation for both original and standardized data
mean_price = np.mean(data_column)
std_dev_price = np.std(data_column)

# Standardize the dataset
standardized_data = (data_column - mean_price) / std_dev_price

# Generate values for the x-axis
x_orig = np.linspace(data_column.min(), data_column.max(), 100)
x_std = np.linspace(standardized_data.min(), standardized_data.max(), 100)

# Plot the original data and the standardized data with normal distribution curves
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(data_column, bins=50, density=True, alpha=0.7, color='blue', label='Original dataset')
pdf_orig = norm.pdf(x_orig, mean_price, std_dev_price)
plt.plot(x_orig, pdf_orig, 'k', linewidth=2, label='Normal Distribution')
plt.title('Original Data')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(standardized_data, bins=50, density=True, alpha=0.7, color='green', label='Standardized dataset')
pdf_std = norm.pdf(x_std, 0, 1)  # Standard normal distribution (mean=0, std=1)
plt.plot(x_std, pdf_std, 'k', linewidth=2, label='Normal Distribution')
plt.title('Standardized Data')
plt.xlabel('Z-Scores')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()
```
